## Expression for potential energy

Hooke’s law states that the force required to stretch or compress a spring is proportionate $\varpropto$ to the distance extended or compressed from its original length. Mathematically, this relation can be expressed as:

$$ F \varpropto x $$
$$ F \varpropto -x $$
$$ F_{r} = -kx  $$

Where,

$F_{r}$ is the force applied to the spring
$x$ is the displacement of the spring
$k$ is a constant which is known as spring constant (stiffness of spring) in units N/m.

Force $F$ can be expressed as negative derivative of potential energy, with respect to a displacement, here in one-dimension as:

$$ F = - \frac{dV}{dx} $$

Let's integrate this function:
To find potential energy within limits 0 $\longrightarrow$ x

$$ \int dV = k \int^{x}_{0} x dx$$
$$ V = \frac{1}{2}kx^{2}$$

Hooke’s law applies to all elastic materials, not just springs. It is an important concept in physics and engineering because it helps to understand and predict the behavior of systems that involve elastic materials, such as springs, rubber bands, and other materials. Hooke’s law is also the basis for the design of many mechanical systems, such as shock absorbers, suspension systems, and other devices that rely on the properties of elastic materials.

# CODE AND PLOT FOR POTENTIAL FUNCTION

The code and plot of potential energy $(V)$ of a particle executing simple harmonic motion against displacement from its equilibrium length is a parabola as illustrated in the following figure.


```python
import matplotlib.pyplot as plt
import numpy as np

k = 1  # spring constant
l = np.linspace(-10, 10, 100)  # range of x values
V = 0.5 * k * l**2  # calculate potential energy

fig, ax = plt.subplots()

# Move the left and bottom spines to x = 0 and y = 0, respectively.
ax.spines[["left", "bottom"]].set_position(("data", 0))

# Hide the top and right spines.
ax.spines[["top", "right"]].set_visible(False)

# Draw arrows (as black triangles: ">k"/"^k") at the end of the axes.  In each
# case, one of the coordinates (0) is a data coordinate (i.e., y = 0 or x = 0,
# respectively) and the other one (1) is an axes coordinate (i.e., at the very
# right/top of the axes).  Also, disable clipping (clip_on=False) as the marker
# actually spills out of the axes.

ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

# annotate
ax.annotate('Turning Point',
            xy=(10, 50),
            xytext=(0.60, 0.90), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.1),
            horizontalalignment='left', verticalalignment='top')

# Plot

plt.plot(l, V)
plt.xlabel(r'$ \longleftarrow \hspace{0.4} \mathrm{Displacement \hspace{0.4} (x)} \hspace{0.4} \longrightarrow$')
plt.ylabel(r'$ \longleftarrow \hspace{0.4} Energy \hspace{0.4} \longrightarrow$')
plt.title(r'$V(x) = \frac{1}{2}kx^2$')
plt.grid()
plt.show()

```

Notice because $V(x)$ is a minimum at $x_{0}$, the first derivative of the potential vanishes there.

Notice also that you can deduce the physically allowable range of motion (energy), and the maximum values of distance and speed, from the limits on the kinetic energy,

$ 0 \leq \textrm{Kinetic Energy (K)} \leq \textrm{Total Energy (E)}$.

Therefore, kinetic energy, $ K=0$, and $ \textrm{Potential Energy V}, V = E$, at the turning point, of which there are two for the elastic spring potential energy,

$$ x_{max} = \pm \sqrt{2E/k}  $$

$ \textrm{called the amplitude} $.

At the bottom of the potential well, $x=0$, $V=0$, and the kinetic energy is a maximum, $K=E=$, so, $\textrm{velocity}_{\textrm{max}} = \pm \sqrt{2E/m}$


# Expression for frequency
The force applied in the section above are usually proportional to the displacement: this is the hallmark of simple harmonic motion. If a particle of mass $m$, which is displaced a distance $x$ from its equilibrium position, experiences a restoring force $-kx$ where $k$ is a constant, its equation of motion and Newton's second law of motion can be expressed like this:

$$ \hat{F} = m \hat{a} $$
$$ ma = -kx$$
$$ m \frac{d^{2}x}{dt^{2}} = -kx$$

Moving terms around I can obtain:

$$\frac{d^{2}x}{dt^{2}} = - \frac{k}{m}x $$

or $$\frac{d^{2}x}{dt^{2}} + \frac{k}{m}x = 0 $$

Solution to this second-order differential equation: $$x(t) = Acos(2\pi t/T + \phi) = Acos(\omega t + \phi)$$

Substitution gives:

$$ -A \omega^{2} cos(\omega t + \phi) = \frac{-k}{m}Acos(\omega t + phi) $$

Cancelling like terms and solving for the angular frequency of the classical oscillator related to the spring constant $k$ in Hooke's law is,

$$ \omega = \sqrt{k/m} $$

Recall, angular frequency is defined as $\omega = 2\pi/T $.

## THE BORN INTERPRETATION

The wavefunction $\Psi$ contains within all the information that can be known about the particle.

Born interpretation of $\Psi$: The probability that a particle will be found in the infinitesimal interval $dx$ about the point $x$, denoted by $P(x) dx$, is

$$ P(x) dx = \left \vert \Psi(x,t) \right \vert^{2} dx   $$

To calculate the probability for a particle to be in the interval \(a < x < b\) of the probability density function $|\Psi|^{2} $ in quantum mechanics and plot the probability density function, you can use the following Python code:

In this code:
1. Define your wave function in the `Psi` function.
2. Specify the interval \([a, b]\) where you want to calculate the probability and visualize using a plot.
3. Use the `spi.quad` function from `scipy.integrate` to perform numerical integration to calculate the probability of the particle being in the interval \([a, b]\).
4. Generate x values for plotting and calculate the corresponding probability density values.
5. Plot the probability density function with shading for the probability interval and dashed lines marking the interval boundaries.

You can run this code in a Python environment to visualize the probability density function and the probability of the particle being in the specified interval.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spi

# Define the probability density function Psi(x)
def Psi(x):
    # Define your wave function here
    return np.exp(-x ** 2)

# Define the interval [a, b]
a = 0
b = 1

# Calculate the probability using numerical integration
probability, error = spi.quad(lambda x: np.abs(Psi(x)) ** 2, a, b)
print(f"The probability of the particle being in the interval ({a}, {b}) is: {probability}")

# Generate x values for plotting
x_values = np.linspace(-3, 3, 1000)
# Calculate the corresponding probability density values
probability_density = np.abs(Psi(x_values)) ** 2

# Plot the probability density function
plt.figure()
plt.plot(x_values, probability_density, label=r'$|\psi(x)|^2$')
plt.fill_between(x_values, probability_density, where=((x_values >= a) & (x_values <= b)), color='skyblue', alpha=0.5, label='Probability Interval')
plt.axvline(x=a, color='red', linestyle='--', label='a')
plt.axvline(x=b, color='red', linestyle='--', label='b')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Density Function')
plt.legend()
plt.grid()
plt.show()
```

# THE HARMONIC OSCILLATOR



The basic time-independent Hamiltonian is: $ H(\hat{x},\hat{p}) = \frac{\hat{p}^{2}}{2m} + \frac{m\omega^{2}\hat{x}^{2}}{2}  $

from above we have that $\omega = \sqrt{k/m}$.

The quantum oscillator is described by the potential energy $V(x) = (1/2)kx^{2} = (1/2)m\omega^{2}x^{2}$ in the Schrodinger equation. After moving terms around we get

$$ \frac{d^{2}\psi(x)}{dx^{2}} = \frac{2m}{\hbar^{2}} \left ( \frac{1}{2} m\omega^{2}x^{2} - E \right ) \psi(x)  $$

as the equation for the stationary states of the oscillator.

Solution is of the form (known as the Gaussian form): $ {\displaystyle \psi(x) = C_{0}e^{-\alpha x^{2}} } $

given here without explanation because it will take more cells to explain. I will make a different notebook just for that lesson.

Taking the second derivative of $\psi(x)$ gives (as you should verify)

$ \frac{d^{2}}{dx^{2}} \psi(x) $ = $ {\displaystyle [4\alpha^{2}x^{2}-2\alpha ] C_{0}e^{-\alpha x^{2}} = [4\alpha^{2}x^{2}-2\alpha ]\psi(x) } $

Comparing this second derived expression with the terms in the Schrodinger equation gives

$ 4\alpha^{2} = \frac{2m}{\hbar^{2}}\frac{1}{2}m\omega^{2} $

where $\alpha$ is determined ,$ \alpha = \frac{m\omega}{2\hbar}$

and

$\frac{2mE}{\hbar^{2}} = 2\alpha = \frac{m\omega}{\hbar}$

where $E$ is determined, $ E = \frac{1}{2}\hbar \omega $

In this way we discover that the oscillator ground-state is described by the wavefunction $ \psi_{0}(x) = C_{0} \textrm{exp}\left ( -{\frac{m\omega x^{2}}{2\hbar} \right )$ and that the energy of this state is $E_{0} = \frac{1}{2}\hbar \omega$.

The constant $C_{0}$ is determined from the normalization condition on the oscillator ground-state wavefunction:

$$ 1 = C_{0}^{2} \int^{\infty}_{-\infty} dx e^{-m\omega x^{2}/\hbar}$$

Evaluation of the integral requires advanced techniques (I will supply these techniques below using quantum operators I learned in intro to QM courses). For now we shall be content here simply to quote the formula (you can verify the formula using Wolfram Mathematica, as I did.)

$$  \int^{\infty}_{-\infty} dx e^{-ax^{2}} = \sqrt{\frac{\pi}{a}}, a > 0 $$

In our case we identify $a$ with $m\omega / \hbar$ and obtain

$$ C_{0}^{2} \sqrt{\frac{\pi \hbar}{m\omega}} $$

Normalization requires this integrated probability to be 1, leading to

$$ C_{0} =  \left ( {\frac{m\omega}{\pi \hbar} \right )^{1/4}  $$

Great!!

We found the ground state eigenfunction as:

$$ \psi_{0} = \left ( {\frac{m\omega}{\pi \hbar} \right )^{1/4}  e^{-m\omega x^{2}/2 \hbar} $$

```python
import matplotlib.pyplot as plt
import numpy as np

# Constants
mass = 1  # Mass of the particle
hbar = 1  # Reduced Planck's constant
omega = 1  # Angular frequency of the harmonic oscillator potential
```

```python
# Potential function for a harmonic oscillator
def potential_energy(x):
    return 0.5 * mass * omega**2 * x**2
```


# GOAL
Our goal is to determine the eigenkets and eigenvalues of the Hamiltonian
$ \hat{H}(\hat{x},\hat{p}) = \frac{\hat{p}^{2}}{2m} + \frac{m\omega^{2}\hat{x}^{2}}{2}  $

Notice the classical Hamiltonian for a harmonic oscillator, with potential $V(x) = \frac{1}{2} kx^{2}$, is a sum of squares. Then there is a simple identity $ C^{*}C = A^{2}+B^{2}$.

Thus, we can write the classical Hamiltonian as:

$$ H = \frac{m\omega^{2}}{2} \left ( x + \frac{ip}{m\omega} \right ) \left ( x - \frac{ip}{m\omega} \right )  $$

# OPERATOR METHODS

Most quantum mechanics textbook use the "ladder operator" method, developed by Paul Dirac. This method allows extraction of the energy eigenvalues without directly solving the differential equation. This formalism is from linear algebra and its application to quantum mechanics. Also, a student will find these operators already defined (versus derived) in quantum textbook.

# By Definition: An OBSERVABLE

An observable is any property of a system that can be measured; e.g. position, momentum, energy, angular momentum, magnetic moment, mass, and so on. In classical mechanics, all observables are functions of a generalized coordinates and momenta, so a knowledge of the physical state implies a knowledge of all possible observables of the system. Basically, one adds or removes fixed quanta of energy from the system.


# By Definition: An OPERATOR

Is a rule for changing functions into other functions. Given any function $ \vert \Psi >$ as input, an operator specifies a unique function $ \vert \Psi^{'} >$ as output.

# Define The Lowering Operator

$$ \hat{a} = \sqrt{\frac{m \omega}{2 \hbar}} \left ( \hat{x} + i \frac{\hat{p}}{m\omega} \right )$$

# Define The Raising Operator

$$ \hat{a}^{\dag} = \sqrt{\frac{m \omega}{2 \hbar}} \left ( \hat{x} - i \frac{\hat{p}}{m\omega} \right )$$

The square root term $ \sqrt{\frac{m \omega}{\hbar}} $ in the raising operator expression encapsulates the physical parameters of the simple harmonic oscillator, including the mass, angular frequency, and Plank's constant. This term governs the scaling of the raising operator with respect to the oscillator's characteristics, ensuring the correct transformation of quantum states.

The term inside the parentheses encapsulates the fundamental properties of the harmonic oscillator.

# Define The Hamiltonian for a Harmonic Oscillator
Rewriting the basic time-independent Hamiltonian is now:

$$ \hat{H} = \frac{1}{2} \hbar \omega \left ( \hat{a}^{\dag} \hat{a} + \hat{a} \hat{a}^{\dag} \right )  $$

We bring in known commutation relations such as:

$ [ \hat{x}, \hat{p} ] = i\hbar  $

and

$ [\hat{a}, \hat{a}^{\dag}] = 1 $

Then

$$ \hat{H} = \hbar \omega \left ( \hat{a}^{\dag} \hat{a} + \frac{1}{2} \right )  $$

Therefore, by definition, these operators have the property of taking an eigenstate of $H$ into another eigenstate, with lower or higher energy.

Let's demo this fact:

# The Wave Function

Earlier we found the ground state eigen function is given by the normalized Gaussian:

$$ \psi_{0} = \left ( {\frac{m\omega}{\pi \hbar} \right )^{1/4}  e^{-m\omega x^{2}/2 \hbar} $$


The state with the next-to-lowest energy is obtained by operating with the raising operator on the ground state.

$$ \psi_{1} = c_{1}\hat{a}^{\dag} \psi_{0}$$

More energy means that for the next-to-lowest energy is:$\\$
$\\$
$ E_{1} = E_{0} + \hbar \omega \\
= \frac{1}{2}\hbar \omega + \hbar \omega \\
= \hbar \omega \left ( 1 + \frac{1}{2} \right )$


In this way we can see that,

$$ \psi_{n} = c_{n}(\hat{a}^{\dag})^{n}\psi_{0} $$

and

$$ E_{n} = \hbar \omega \left ( n + \frac{1}{2} \right ) $$

for $n = 0,1,2,3,...$

but we now know:

$ c_{n} = \frac{1}{\sqrt{n!}}$

then to find the higher energy states, consider

$$ \psi_{n}(x) = \frac{1}{\sqrt{n!}}  ( \hat{a}^{\dag})^{n} \psi_{0}(x)  $$


set $(\hat{a}^{\dag})^{n}$ to
$$ \hat{a}^{\dag} = \sqrt{\frac{m \omega}{2 \hbar}} \left ( \hat{x} - i \frac{\hat{p}}{m\omega} \right )$$

thus,

$$ \psi_{n}(x) = \frac{1}{\sqrt{n!}}  \left ( \sqrt{\frac{m \omega}{2 \hbar}} \left ( \hat{x} - i \frac{\hat{p}}{m\omega} \right ) \right )^{n} \psi_{0}(x)  $$


$$ \psi_{n}(x) = \frac{1}{\sqrt{n!}} \left (  \sqrt{\frac{m\omega}{2\hbar}} \right )^{n} \left ( x -  \frac{\hbar}{m\omega} \frac{\partial{}}{\partial{x}}\right )^{n} \psi_{0}(x)  $$

# SPECIAL FUNCTIONS
# HERMITE POLYNOMIALS

Hermite function is the solution of second-order differential equation has the form

$$ \frac{d^{2}y}{dt^{2}} - 2x \frac{dy}{dx} + 2ny = 0   $$


We define a quantity $ \chi = \sqrt{\frac{m\omega}{\hbar}} x$

and
$\\$

$ \frac{d\chi}{dx} = \left ( \frac{m\omega}{\hbar} \right )^{1/2}  $

therefore,

$$ \psi_{n}(\chi) = \frac{1}{\sqrt{n!}} * \left ( \frac{m\omega}{\pi \hbar} \right )^{0.25} * \left ( \frac{1}{\sqrt{2}} \right )^{n} * \left ( \chi - \frac{\partial }{\partial \chi} \right )^{n} * e^{-\chi^{2}/2}  $$

$$ \psi_{n}(\chi) = \frac{1}{\sqrt{n!}} * \left ( \frac{m\omega}{\pi \hbar} \right )^{0.25} * \left ( \frac{1}{\sqrt{2}} \right )^{n} *  \mathcal{H}_{n}(\chi) * e^{-\chi^{2}/2}  $$

# with the corresponding eigenvalues

$$ E_{n} = \hbar \omega \left ( n + \frac{1}{2} \right )$$

The function $ \mathcal{H}_{n}(\chi)$ are known as $ \textbf{Hermite Polynomials} $.

$\\$

# CODE AND PLOT

```python
#  Define the wave function for the quantum harmonic oscillator
def wave_function(x, n):
    prefactor = 1 / np.sqrt(2**n * np.math.factorial(n)) * (mass * omega / (np.pi * hbar))**0.25
    exponent = -0.5 * mass * omega * x**2 / hbar
    hermite_polynomial = np.polynomial.hermite.hermval(np.sqrt(mass * omega / hbar) * x, [0] * (n + 1) + [1])
    return prefactor * hermite_polynomial * np.exp(exponent)
```

# USING THE NUMPY HERMVAL FUNCTION

Evaluate an Hermite series at point x.

if $c$ is of length $n+1$, this function returns the value

$$ p(\chi) = c_{0} * \mathcal{H}_{0}(\chi) + c_{1} * \mathcal{H}_{1}(\chi) + ... + c_{n} *\mathcal{H}_{n}(\chi) $$

The parameter $\chi$ is converted to an array only if it is a tuple or a list, otherwise it is treated as a scalar. In either case, either $\chi$ or its elements must support multiplication and addition both with themselves and with the elements of $c$.

If $c$ is a 1-D array, then $p(\chi)$ will have the same shape as $\chi$. If $c$ is multidimensional, then the shape of the result depends on the value of tensor. If tensor is true the shape will be c.shape[1:] + x.shape. If tensor is false the shape will be c.shape[1:]. Note that scalars have shape (,).

Trailing zeros in the coefficients will be used in the evaluation, so they should be avoided if efficiency is a concern.


source: https://numpy.org/doc/stable/reference/generated/numpy.polynomial.hermite.hermval.html

```python
# Define the potential energy curve for plotting
x_vals = np.linspace(-5, 5, 1000)
potential_vals = potential_energy(x_vals)
```

```python
# Plot the potential energy curve
plt.figure()
plt.plot(x_vals, potential_vals, label='Potential Energy')
plt.xlabel('x')
plt.ylabel('V(x)')
plt.title('Harmonic Oscillator Potential Energy')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plot the wave functions for the quantum harmonic oscillator
plt.figure()
for n in range(3):
    wave_func_vals = wave_function(x_vals, n)
    plt.plot(x_vals, wave_func_vals, label=f'n = {n}')
plt.xlabel('x')
plt.ylabel('Wave Function')
plt.title('1-Dim Wave Functions for the Harmonic Oscillator')
plt.legend()
plt.grid()
plt.show()
```
