### SECTION 5.7 Planck's Blackbody Radiation Formula

$\textbf{Defintions and Equations:} $ 

Although absolute thermodynamic equilibrium is rare, it is useful to employ the concept of local thermodynamic equilibrium (LTE). In LTE, the radiation field is isotropic, so we can write $I_{ν} = J_{ν}$

$J_{v}$ is denoted as the mean specific intensity.

And with the photons that have frequency : 

$ v = \frac{\Delta E}{h}$

or re-written as

$ hv = \Delta E$

We say,

The radiation that is emitted from a small opening in the cavity is described by Planck’s law, which gives the spectral radiance as a function of the temperature T and frequency ν, or wavelength λ:


$$ I(v,T) = I_{v} (T) = \frac{2hv^{3}}{c^{2}} \frac{1}{e^{hv/kT} - 1} $$

with $ h = 6.626 * 10^{-34} $ J s

and units: $ \frac{W}{m^{2}srHz}$


$$ \textbf{equation 5.86}$$ 

$$ I_{\lambda} = \frac{2hc^{2}}{\lambda^{5}} * \frac{1}{e^{hc/\lambda kT} -1} $$

and units: $ \frac{W}{m^{3}sr}$

$$ \textbf{equation 5.90}$$ 

c is the speed of light (m/s), h is Plank's Constant (Js), $\lambda$ is the wavelength (m), k is Boltzman's Constant (J/K), T is temperature (K), and $\epsilon = 1 $ in the numerator.

# Build model using frequency $v$

- For the Planck function $I_{ν} (T )$ (see equation 5.86), what is the most probable frequency $ν_{p}$ at a given temperature T ? 

Use this function to plot emission spectra for stars at 4000K and 5000K (the surface of the Sun is about 5800 K).

Note from above:


$$ I(v,T) = I_{v} (T) = \frac{2hv^{3}}{c^{2}} \frac{1}{e^{hv/kT} - 1} $$

$ v = \frac{c}{\lambda}$

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import pandas as pd 
%matplotlib inline
```

# Write a function in terms of frequency and temperature
# Call the function Inu (nu, T), to return $ I_{v} (v,T) = \frac{2hv^{3}}{c^{2}} \frac{1}{e^{hv/kT} -1} $ 

```python

def Inu(nu, T):
    """Calculates the radiance from frequency and temperature.

    Arguments
    nu: frequency Hz
    T: temperature K
    """

    return  (2 * h * nu**3 / c**2) * (1 / (np.exp(h*nu/(kb*T)) - 1))

nu1 = c/200e-9
nu2 = c/900e-8     # frequency


x = np.linspace(nu1,nu2,1000)
y = Inu(x,4000)

plt.figure()
plt.plot(x,Inu(x,4000),'g')
plt.plot(x,Inu(x,5000),'r')
plt.plot(x,Inu(x,7000),'k')  
plt.xlabel('Frequency - Hz')
plt.ylabel('Intensity - Watts')# Intensity versus frequency

print('Black represents Betelgeuse, Red - Sun, Green-Sirius')
```

# Build model using Wavelength $ \lambda $

- For the Planck function expressed as a function of wavelength, $I_{λ}(T)$ (see equation 5.90), what is the most probable wavelength $λ_{p}$ at a given temperature T ? 

$$ I_{\lambda} = \frac{2hc^{2}}{\lambda^{5}} * \frac{1}{e^{hc/\lambda kT} -1} $$

and units: $ \frac{W}{m^{3}sr}$

# Write a function in terms of wavelength and temperature
# Call the function  rho( wav, T), to return $ I_{\lambda} $ 

```python
# package utility

from scipy.constants import  Planck, speed_of_light, Boltzmann

# define common constants

h = Planck            # Js
c = speed_of_light    # m/s
kb = Boltzmann        # J/molec K


def rho(wav, T):
    """Calculate the emission from wavelength and temperature.

    Arguments::
    lamda: wavelength cm
    T: temperature K"""

    return 2 * h * c**2 / ( wav**5 * (np.exp(h*c/(wav*kb*T)) - 1))
```

# between ultraviolet and infrared spectrum of 100 nm to 3000 nm

```python
wav =np.linspace(1*10**-7, 3*10**-6,1000)
```

```python
TK = 4000   # temperature of a K-type star
TG = 5500   # temperature of a G-type star
TF = 7000   # temperature of a F-type star
IntensityK = rho(wav,TK)
IntensityG = rho(wav,TG)
IntensityF = rho(wav,TF)

plt.figure()
plt.plot(wav,IntensityK,'g-')
plt.plot(wav, IntensityG,'k-')
plt.plot(wav, IntensityF, 'r-')
plt.xlabel('Wavelength - nm')
plt.ylabel('Intensity - Watts')# Intensity versus frequency
```
