# Understanding Radioactive Decay

Radioactive decay, also referred to as nuclear decay or radioactivity, is the process by which an unstable atomic nucleus releases energy in the form of radiation. A substance with unstable nuclei is classified as radioactive. The three primary types of decay are alpha, beta, and gamma decay. Beta decay occurs due to the weak force, while alpha and gamma decays are influenced by electromagnetic and nuclear forces, respectively.

It's important to note that radioactive decay is inherently random when observed at the level of individual atoms. According to quantum theory, it's impossible to predict exactly when a specific atom will decay, regardless of its age. However, when considering a large number of identical atoms, the overall decay rate can be quantified using a decay constant or a half-life. The half-lives of radioactive materials can vary dramatically, ranging from almost instantaneous to durations exceeding the age of the universe itself.

# Exponential Decay

Below is a Python code that models the radiation decay of a radioactive substance, americium-241, using the exponential decay law. The half-life of americium-241 is approximately 432 years. This code simulates how the quantity of americium-241 decays over time.

I have for you an explanation of decay (at the bottom). Given that americium-241 decays approximately 432 years, the decay constant, $ \lambda $, of Americium-241 represents the probability per unit time that a single atom of Americium-241 will undergo radioactive decay, essentially indicating how quickly a given sample of Americium-241 will decrease in quantity over time due to its radioactive nature; a higher decay constant means a faster decay rate, and since Americium-241 has a long half-life, its decay constant is relatively small, meaning it decays very slowly.


$ \lambda = \frac{\ln(2)}{432} = 0.001605 = 1.6 *  10^{-3}$ per year.

KEY POINTS:
- It primarily decays by emitting alpha particles.
- Commonly used in smoke detectors due to its alpha radiation.
- Due to its long half-life, the decay rate is relatively slow.

THE MODEL:

I assume the initial decay quantity with with arbitrary units. So this can mean, 1000 grams, or 1000 atoms.

If it is 1000 atoms, after 432 years say approximately 500 atoms will have decayed. For 1000 grams, you can see from the plot, approximately 500 grams of americium-241 will decay in 432 years.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters for americium-241
decay_constant = np.log(2) / 432  # Decay constant (lambda) in 1/year
initial_quantity = 1000  # Initial quantity of americium (in arbitrary units)
time_period = 5000  # Time period for the simulation (5000 years)
time_steps = np.linspace(0, time_period, 1000)  # Time steps
```

```python
# Calculate remaining quantity over time
remaining_quantity = initial_quantity * np.exp(-decay_constant * time_steps)
```

```python
# Visualization
plt.figure(figsize=(12, 8))
plt.plot(time_steps, remaining_quantity, label='Americium-241 Decay', color='blue')
plt.title('Radioactive Decay of Americium-241 Over Time')
plt.xlabel('Time (years)')
plt.ylabel('Remaining Quantity (arbitrary units)')
plt.axhline(y=0, color='gray', linestyle='--')
plt.axvline(x=0, color='gray', linestyle='--')
plt.ylim(0, initial_quantity * 1.1)  # Set y-limit for better visibility
plt.xlim(0, time_period)  # Set x-limit to 5000 years
plt.legend()
plt.grid()
plt.show()
```

### Explanation of Exponential Decay

A decaying quantity, such as $ N(t) $, can change over time $\frac{d}{dt}$ if it decreases at a rate $\lambda$ proportional to its current value. Mathematically this can be written as,

$\frac{dN(t)}{dt} = - \lambda N(t)$

The Mean-lifetime is defines as $\tau = \frac{1}{\lambda}$, solving for lambda, $ \lambda = \frac{1}{\tau}$

The Half-life is defined $ t_{1/2} = \frac{\ln{2}}{\lambda} $ solving for lambda, $\lambda$ $= \frac{\ln{2}}{t_{1/2}}$ or in terms of the mean lifetime $ t_{1/2} = \tau \ln{2}$.

1. **Decay Constant**: The decay constant is calculated using the formula $ \lambda = \frac{\ln(2)}{\tau_{1/2}} $, where $ \tau_{1/2} $ is the half-life of americium-241 (432 years).
2. **Parameters**:
   - `initial_quantity`: The initial amount of americium, set to 1000 (arbitrary units).
   - `time_period`: The total duration for the simulation (set to 5000 years).
3. **Remaining Quantity Calculation**: Utilizing the exponential decay formula $ N(t) = N_0 e^{-\lambda t} $ to compute the remaining quantity over time.
4. **Visualization**: The Matplotlib library is used to generate a plot of the decay curve, where the x-axis indicates time (in years) and the y-axis indicates the remaining quantity of americium.
